### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total_score'] / 3
```

```python
df.head()
```

```python
math_full = df[df['math_score']==100]['average'].count()
reading_full = df[df['reading_score']==100]['average'].count()
writing_full = df[df['writing_score']==100]['average'].count()

print(f"Number of student with full marks in Math: {math_full}")
print(f"Number of student with full marks in Reading: {reading_full}")
print(f"Number of student with full marks in Writing: {writing_full}")
```

```python
math_less_20 = df[df['math_score']<=20]['average'].count()
reading_less_20 = df[df['reading_score']<=20]['average'].count()
writing_less_20 = df[df['writing_score']<=20]['average'].count()

print(f"Number of student with less than 20 marks in Math: {math_less_20}")
print(f"Number of student with less than 20 marks in Reading: {reading_less_20}")
print(f"Number of student with less than 20 marks in Writing: {writing_less_20}")
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### 4.4.1 GENDER COLUMN
- How is distribution of Gender ?
- Is gender has any impact on student's performance?

#### UNIVARIATE ANALYSIS ( How is distribution of Gender ? )

```python
f,ax=plt.subplots(1,2,figsize=(15,6))
sns.countplot(x=df['gender'],data=df,palette='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=15)
    
ax[1].pie(x=df['gender'].value_counts().values,labels=df['gender'].value_counts().index,explode=[0,0.05],autopct='%.2f%%',shadow=True,colors=['red','green'])
plt.show()
```

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender')[['average','math_score']].mean()
gender_group
```

```python
plt.figure(figsize=(10, 6))

X = ['Total Average','Math Average']


female_scores = [gender_group['average'][0], gender_group['math_score'][0]]
male_scores = [gender_group['average'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.

#### 4.4.4 LUNCH COLUMN 
- Which type of lunch is most common amoung students ?
- What is the effect of lunch type on test results?

#### UNIVARIATE ANALYSIS ( Which type of lunch is most common amoung students ? )

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
sns.countplot(data=df,x=df['lunch'],palette='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=10)

ax[1].pie(x=df['lunch'].value_counts().values,labels=df['lunch'].value_counts().index,explode=[0.003,0],autopct='%.2F%%',shadow=True)
plt.show()
```

#### BIVARIATE ANALYSIS (  Is lunch type intake has any impact on student's performance ? )

```python
lunch_groupby = df.groupby('lunch')[['total_score','average']].mean()
lunch_groupby
```

```python
plt.figure(figsize=(10, 6))

X = ['Total Average','Average']

free_lunch_score = [lunch_groupby['average'][0], lunch_groupby['total_score'][0]]
standad_lunch_score = [lunch_groupby['average'][1], lunch_groupby['total_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, free_lunch_score, 0.4, label = 'Free/reduced Lunch')
plt.bar(X_axis + 0.2, standad_lunch_score, 0.4, label = 'Standard Lunch')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Average marks of both the different lunch", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- Students who get Standard Lunch tend to perform better than students who got free/reduced lunch
